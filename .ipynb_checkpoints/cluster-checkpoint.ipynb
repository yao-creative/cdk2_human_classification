# Clustering


```python
import os
import numpy as np
import matplotlib
import pickle
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```python
with open("matrix.var", "rb") as matrix_var:
   matrix = pickle.load(matrix_var) 
print(f"matrix: {matrix}")
with open("chains_list.var","rb") as chains_list_var:
   chains_list = pickle.load(chains_list_var)
print(f"chains_list: {chains_list}")
```

```python
out = linkage(matrix, 'ward')
fig = plt.figure(figsize=(50,20))
dn = dendrogram(out)
print(f"out: {out}")
```

```python
out = linkage(matrix, 'single')
fig = plt.figure(figsize=(50,20))
dn = dendrogram(out)
```

```python
out = linkage(matrix, 'average')
fig = plt.figure(figsize=(500,200))
dn = dendrogram(out)
```

```python
out = linkage(matrix, 'ward',optimal_ordering=True)
fig = plt.figure(figsize=(50,20))
dn = dendrogram(out)
```
