# Finding High Variance Regions on the chain
Use this method to identify crucial subsequences which can be used to identify and cluster the conformations

```python
from helper import identify_highvariance_region, pick_hv_regions
import pickle
import matplotlib.pyplot as plt 
from pandas import DataFrame as df
import numpy as np
```

```python
with open("coords_res.var", "rb") as coords_res_var:
    coords_res = pickle.load(coords_res_var)
conformation_by_index_link_var = np.array(identify_highvariance_region(coords_res))
```

```python
y = conformation_by_index_link_var
x = [i if i%5 == 0 else None for i in np.arange(len(conformation_by_index_link_var)) ]
#print(f"x: {x}")
heat_frame = df(np.matrix([y]))

fig = plt.figure(figsize =(10,5))
plt.imshow(heat_frame, cmap='autumn',interpolation="nearest",aspect=8, vmax = max(conformation_by_index_link_var), vmin= min(conformation_by_index_link_var))
plt.xticks(range(len(x)),x,fontsize=120)
plt.title("Inter residue vector average coordinate variance heat map")
fig.subplots_adjust(left=0.1, right=30.0, top=20.0, bottom=0.1)
plt.show()
```

```python
y = conformation_by_index_link_var
x = np.arange(0,len(conformation_by_index_link_var))#[i if i%5 == 0 else None for i in np.arange(len(conformation_by_index_link_var)) ]
#print(f"x: {x}")
fig = plt.figure(figsize =(10,5))
plt.plot(x,y, label = "variance")
plt.title("Inter residue vector average coordinate variance plot")
```

It is obvious to see that the variance in the region 34-51 and 45-64 are especially high. There's some mall noticable regions 11-15 and the 293-295 and some variations on specific isolated residues like #72.
